```python
# ==================================
# Raw vs Preprocessed Comparison EDA
# ==================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Load datasets ---
df_raw = pd.read_csv("tamil_nadu_rural_bus_dataset_enriched_100k.csv")
df_processed = pd.read_csv("tamil_nadu_bus_sample_enriched_processed.csv")
```

```python
# =============================
# 1. Compare Missing Values
# =============================
missing_raw = df_raw.isnull().sum()
missing_processed = df_processed.isnull().sum()

plt.figure(figsize=(10,5))
sns.barplot(x=missing_raw.index, y=missing_raw.values, color="red", alpha=0.6, label="Raw")
sns.barplot(x=missing_processed.index, y=missing_processed.values, color="green", alpha=0.6, label="Processed")
plt.xticks(rotation=90)
plt.ylabel("Missing Value Count")
plt.title("Missing Values: Raw vs Processed")
plt.legend()
plt.show()
```

```python
# =============================
# 2. Distribution of Target (delay_minutes)
# =============================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df_raw['delay_minutes'], kde=True, bins=30, color="red")
plt.title("Raw Dataset: Delay Minutes Distribution")

plt.subplot(1,2,2)
sns.histplot(df_processed['delay_minutes'], kde=True, bins=30, color="green")
plt.title("Processed Dataset: Delay Minutes Distribution")
plt.show()
```

```python
# =============================
# 3. Distribution of Traffic Level
# =============================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.countplot(data=df_raw, x="traffic_level", order=df_raw['traffic_level'].value_counts().index, palette="Reds")
plt.title("Raw Dataset: Traffic Level")

plt.subplot(1,2,2)
sns.countplot(data=df_processed, x="traffic_level", order=df_processed['traffic_level'].value_counts().index, palette="Greens")
plt.title("Processed Dataset: Traffic Level")
plt.show()
```

```python
# =============================
# 4. Boxplot: Delay vs Weather
# =============================
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(data=df_raw, x="weather", y="delay_minutes", palette="Reds")
plt.title("Raw Dataset: Delay vs Weather")
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.boxplot(data=df_processed, x="weather", y="delay_minutes", palette="Greens")
plt.title("Processed Dataset: Delay vs Weather")
plt.xticks(rotation=45)
plt.show()
```

```python
# =============================
# 5. Correlation Heatmap
# =============================
import numpy as np

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.heatmap(df_raw.select_dtypes(include=np.number).corr(), cmap="coolwarm", cbar=False)
plt.title("Raw Dataset: Correlation Heatmap")

plt.subplot(1,2,2)
sns.heatmap(df_processed.select_dtypes(include=np.number).corr(), cmap="coolwarm", cbar=False)
plt.title("Processed Dataset: Correlation Heatmap")
plt.show()
```
